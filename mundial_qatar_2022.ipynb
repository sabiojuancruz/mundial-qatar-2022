```python
# Cuantas fases tiene el mundial?
# Completar
fases = mundial['group'].unique().tolist()
for i in range(len(fases) - 1, -1, -1):
  if "Group" in fases[i]:
    fases.pop(i)

fases.append("Group-Fase")

len(fases)
```

```python
# Cual fue la "attendance" promedio por partido?
# Completar,
float(mundial["attendance"].mean())
```

```python
# Cual fue el partido con mas diferencia de goles?
# Completar
goles1 = mundial["1_goals"]
goles2 = mundial["2_goals"]
dif_goles = abs(goles1 - goles2)

index_max = dif_goles.idxmax()
mundial.loc[index_max]
```

```python
# Cual es la selecciÃ³n que mas penales metiÃ³?
# Completar
penales1 = mundial["1_panelties_scored"]
penales2 = mundial["2_panelties_scored"]
equipos1 = mundial["1"]
equipos2 = mundial["2"]

penales_equipos1 = pd.concat([equipos1, penales1], axis=1).groupby(by="1").sum()
penales_equipos2 = pd.concat([equipos2, penales2], axis=1).rename(columns = {'2':'1'}).groupby(by="1").sum()

penales_equipos = pd.merge(penales_equipos1, penales_equipos2, how='inner', on='1').groupby(by="1").sum()
penales_totales = penales_equipos['1_panelties_scored'] + penales_equipos['2_panelties_scored']

mas_penales = penales_totales.idxmax()
mas_penales
```

```python
# Cual es la selecciÃ³n que mas penales le hicieron?
# Completar
penales1 = mundial["1_panelties_scored"]
penales2 = mundial["2_panelties_scored"]
equipos1 = mundial["1"]
equipos2 = mundial["2"]

penales_en_contra_equipos1 = pd.concat([equipos1, penales2], axis=1).groupby(by="1").sum()
penales_en_contra_equipos2 = pd.concat([equipos2, penales1], axis=1).rename(columns = {'2':'1'}).groupby(by="1").sum()

penales_en_contra_equipos = pd.merge(penales_en_contra_equipos1, penales_en_contra_equipos2, how='inner', on='1').groupby(by="1").sum()
penales_totales = penales_en_contra_equipos['1_panelties_scored'] + penales_en_contra_equipos['2_panelties_scored']

mas_penales_en_contra = penales_totales.idxmax()
mas_penales_en_contra
```

### Segunda Parte

```python
equipo : str = "ARGENTINA"
```

```python
# Seleccionar todos los partidos que jugo el equipo de "local"
# Completar

filtroLocal : pd.Series = mundial.loc[mundial['1'] == equipo]
filtroLocal
```

```python
# Seleccionar todos los partidos que jugo el equipo de "visitante"
# Completar

filtroVisitante : pd.Series = mundial.loc[mundial['2'] == equipo]
filtroVisitante
```

```python
# Seleccionar todos los partidos que jugo el equipo (de "local" o "visitante")
# Completar

partidos : pd.DataFrame = mundial.loc[(mundial['1'] == equipo) | (mundial['2'] == equipo)]
partidos
```

```python
# Cuantos goles hizo el equipo en total (obtener los goles metidos de local y sumarlos con los goles metidos de visitante)
# Completar

goles : int = int(filtroLocal['1_goals'].sum() + filtroVisitante['2_goals'].sum())
goles
```

```python
# Cuantos goles hizo el equipo en promedio
# Completar
promedio : float = goles / len(partidos)
promedio
```

```python
# Cuantos partidos gano el equipo
# Completar

ganados : int = len(filtroLocal.loc[filtroLocal['1_goals'] > filtroLocal["2_goals"]]) + len(filtroVisitante.loc[filtroVisitante['2_goals'] > filtroVisitante["1_goals"]])
ganados
```

```python
# Graficar los goles hechos por el equipo en cada partido

fig, ax = plt.subplots()
# Completar
```
